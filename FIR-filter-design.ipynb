```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# A plot function
def plotter(x_label,y_label,title,x_axis,y_axis):
    plt.figure(figsize=(12,8))
    plt.plot(x_axis,y_axis)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True,which = 'both')
    plt.axhline(y=0, color='k')
    
```

```python
# getting x-values of sine wave 
time = np.arange(0,10,0.1)
time
```

```python
# amplitude of the sine wave
amplitude = np.sin(time)
amplitude
```

```python
# plotting the sine wave for validation
plotter(title= 'Sine Wave', x_label= 'Amplitude', y_label='Time',x_axis= time,y_axis= amplitude)
```

```python
# Lets add some noise to the data
noise = np.random.normal(0,1,100)
noisy_signal = amplitude + noise
plotter(x_axis=time,y_axis=noisy_signal,x_label='Time',y_label='Amplitude of Noisy Signal',title='Noisy Signal')
```
